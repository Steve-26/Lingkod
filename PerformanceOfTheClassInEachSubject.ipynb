```python
data['math_PassStatus'] = np.where(data['math score']<40, 'F', 'P')
```

```python
data['read_PassStatus'] = np.where(data['reading score']<40, 'F', 'P')
```

```python
data['write_PassStatus'] = np.where(data['writing score']<40, 'F', 'P')
```
